### Data Curation, Parsing, and Management

In this section I will demonstrate how to retrieve the dataset off the internet, load the data into the jupyter environment, and wrangle with the data to tidy it in order to prepare for analysis.

```python
# Begin by importing the necessary modules

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import zipfile
from io import BytesIO
```

In this section, I will download a zip file hosted on github that contains the dataset. I then will extract the files out of the zipfile into the current directory. Once extracted, I load the .csv files into the environment.

```python
# Define the url to the .zip file
url = 'https://github.com/chadwickbureau/baseballdatabank/archive/master.zip'

# Downloading the file with the requests library
req = requests.get(url)

# Extracting the dataset from the downloaded zip file
zipfile= zipfile.ZipFile(BytesIO(req.content))
zipfile.extractall('./')

# Loading the .csv files necessary for analysis into pandas dataframes
df_batting = pd.read_csv("baseballdatabank-master/core/Batting.csv")
df_fielding = pd.read_csv("baseballdatabank-master/core/Fielding.csv")
df_pitching = pd.read_csv("baseballdatabank-master/core/Pitching.csv")
df_hof = pd.read_csv("baseballdatabank-master/core/HallOfFame.csv")
df_people = pd.read_csv("baseballdatabank-master/core/People.csv")
```

Now that the data has been downloaded and is housed in a pandas dataframe, it is time to wrangle the data to fit our needs. For the purpose of formalizing Hall of Fame criteria the goal is to compose a table that houses all relevant career stats for each player throughout history. The dataset structure separates various types of stats into different tables. For instance, batting, fielding, and pitching each have their own table. We will want to combine these to represent a players career in one row, all in one table.
<br>
<br>
I have chosen to separate players into two distinct groups: Pitchers and Position Players. Historically in baseball there are exceptionally few players to ever pitch and play a non-pitcher position. Even fewer players to ever do this with any sort of consistency. It makes most sense to separate these players and treat them differently. They are also treated differently by pundits and have entirely different criteria for evaluation.
<br>
<br>
We also will want to add the column "HoF" for each player which will be a boolean value "True" or "False" that will define whether the player is a Hall of Famer or not. This will be necessary for the machine learning section where a classifier is built on labeled training data.
<br>
<br>
This is the initial state of the dataframe as given by the .csv file. 

```python
df_batting.head()
```

In the initial state shown above, each row represents one player's batting stats for one year. I wish to condense this table into each row representing one player's career statistics. To do this I will need to take every row for each playerID and sum the stats into one cumulative row for their career. This is valid because all statistics are numerical values where summing is valid.

```python
# Removing columns that are unnecessary for my analysis to limit clutter.
df_batting_dropped = df_batting.drop(columns = ["yearID", "stint", "teamID", "lgID"])

# This line groups by playerID and then sums the columns of each group. The effect of this is taking
# all rows with equal playerIDs and combining them into one, cumulative row for the player's career.
df_batting_career = df_batting_dropped.groupby(["playerID"], as_index=False).sum()

# Later in the fielding section, there are duplicate column names for "G", "SB", and "CS" but
# they mean different things. I rename them to "G_b" for G_batting, etc. for clarity.
df_batting_career = df_batting_career.rename(columns={"G": "G_b", "SB": "SB_b", "CS": "CS_b"})

# Here is the result of this stage, we now have career batting stats in each row!
df_batting_career.head()
```

In this section I will perform the process shown above, but on fielding data. The structure of the data and method for wrangling is very similar.

```python
# Initial state
df_fielding.head()
```

One uniqueness to fielding data that poses a new challenge is that there is a new column "POS" that describes the fielders position. This is problematic because we cannot "sum" qualitative position values over a career in any meaningful way. I chose to handle this by assigning each player one career position, their most popular. This is likely how players are remembered and defined. I do this by checking the players most played position by Games Played.

```python
# Dropping unnecessary columns
df_fielding_dropped = df_fielding.drop(columns = ["yearID", "stint", "teamID", "lgID"])

# This groups by playerID as seen before, but now we also group by position so that we obtain
# career stats for each player and each position. I then sort these values by playerID, then
# Games. The effect of this is having the first row of each player being their most played position.
df_fielding_career_by_pos = df_fielding_dropped.groupby(["playerID", "POS"], as_index=False).sum() \
                                        .sort_values(['playerID', 'G'], ascending=[True, False])

# The resulting table.
df_fielding_career_by_pos.head()
```

The purpose of this following section is to select each player's most played defensive position. I do this by doing another groupby playerID but this time instead of summing all columns, I will sum all columns except for "POS". For position I define the function to aggregate with to be "first". This will select the position value first seen, which will be there most played position as setup above.

```python
# Define a dictionary of aggregation functions
# This maps from column names to aggregation function. It defines all columns to be summed
f = dict.fromkeys(df_fielding_career_by_pos, 'sum')

# Explicitly remove playerID because this is the variable we are grouping on
f.pop("playerID")

# Explicitly define POS to be aggregated with the "first" function
f["POS"] = "first"

# Carry out grouping and aggregation using f
df_fielding_career = df_fielding_career_by_pos.groupby(["playerID"], as_index=False).agg(f)

# Rename columns as described in the batting section
df_fielding_career = df_fielding_career.rename(columns={"G": "G_f", "SB": "SB_f", "CS": "CS_f"})

# The resulting career stats table!
df_fielding_career.head()
```

In this section we will merge the tables for career batting and fielding stats to create one all-encompassing table for a player's career. I choose to merge with an outer join in order to preserve any players that may have only batting or only fielding stats. We would not want these players to be removed from our table.

```python
# Merge the two career tables with an outer join on playerID
df_batting_and_fielding_career = df_batting_career.merge(df_fielding_career, how='outer', on='playerID')
df_batting_and_fielding_career.head()
```

One thing we must now do to allow this table to be useful for position players is to remove pitchers! The pitchers are all still in this table because they have fielding or batting statistics, but we do not want them in the same table as position players since we have decided to separate them.

```python
df_batting_and_fielding_career = df_batting_and_fielding_career[df_batting_and_fielding_career['POS'] != 'P']
```

As the last step of the wrangling process we must define whether a player has been inducted into the MLB Hall of Fame for the sake of machine learning and classification. I do this by creating a dictionary that maps every MLB player in history to a boolean that will be "True" if they are a Hall of Famer and "False" if they are not.

```python
# Initialize dict
is_hof = dict()

# Set all values to False initially, to be revised below
for player in df_batting_and_fielding_career['playerID']:
    is_hof[player] = 0
    
# Loop through the Hall of Fame table from our dataset to obtain info on HoF Induction status.
for row in df_hof.iterrows():
    # If the player has been inducted AND the player is in our dictionary, set their value to true.
    # The second condition of being in the dict is necessary because not all Hall of Fame inductees
    # are players that we care about for our analysis. Some inductees are managers, executives, or
    # other non-player roles in baseball.
    if row[1]['inducted'] == 'Y' and row[1]["playerID"] in is_hof.keys() and row[1]['category'] == 'Player':
        is_hof[row[1]["playerID"]] = 1
         
# Create a 'HoF' column that is set to the values of our dictionary.
# The tables will align because the dictionary was initialized in the ordering
# of 'playerID' in the career stats table.
df_batting_and_fielding_career['HoF'] = is_hof.values()

# Our resulting table with HoF values!
df_batting_and_fielding_career.head()
```

We have now created the table for position players! Great, now it is time to create the table for pitchers. This will involve similar processes but will be notably easier because all stats are found in one table rather than being divided between two tables for batting and fielding as for position players.
<br>
<br>
We begin by viewing the initial state of the table.

```python
df_pitching.head()
```

I begin by removing unnecessary columns and summing career statistics for each player as we did for position players.

```python
# Removing columns that are unnecessary for my analysis to limit clutter.
df_pitching_dropped = df_pitching.drop(columns = ["yearID", "stint", "teamID", "lgID"])

df_pitching_career = df_pitching_dropped.groupby(["playerID"], as_index=False).sum()

df_pitching_career.head()
```

We now have career stats for all the pitchers! All that's left to do is to add a HoF column in the same way we did before.

```python
# Initialize dict
is_hof = dict()

for player in df_pitching_career['playerID']:
    is_hof[player] = 0
    
for row in df_hof.iterrows():
    if row[1]['inducted'] == 'Y' and row[1]["playerID"] in is_hof.keys() and row[1]['category'] == 'Player':
        is_hof[row[1]["playerID"]] = 1

df_pitching_career['HoF'] = is_hof.values()

# Our resulting table with HoF values!
df_pitching_career.head()
```

### Exploratory Data Analysis
In this section we will do some initial exploratory analysis of the data we've cleaned in order to investigate suspected relationships between the data. We will use some libraries to assist with regression as well as data visualization. Visualizing the data is a great way to help us with EDA as we can spot relationships in the data much easier in well composed graphs than in very large tables.
<br>
<br>
Let's begin our investigation into what we think may cause players to be thought of as Hall of Fame quality. Players who are *consistently* good are much more likely to be inducted than those who are exceptional for a short period of time. For this reason I believe time in the league is a very important quality. Let's begin by exploring how games played (position players) and innings pitched (pitchers) effect hall of fame induction.
<br>
<br>
The strategy I'm choosing to use will be to calculate the percentage of players who are Hall of Famers in certain ranges of Games Played. The bins will be of width 300 so I will calculate how what percentage of players with 0-300 games played, 300-600 games played etc. have been inducted into the Hall of Fame.

```python
# Array to hold the HoF percent per 'bin'
pct_of_hof = []

# Used to define the range of values for bins
max_games = df_batting_and_fielding_career['G_b'].max()

# For each bin (defining width of 300 games)
for i in range(0, max_games, 300):
    # Filter to dataframe to include only players with games played within the range for this bin
    bin = df_batting_and_fielding_career[(df_batting_and_fielding_career['G_b'] > i) 
                                       & (df_batting_and_fielding_career['G_b'] <= i + 300)]
    
    # The total number of players in this bin
    num_players = bin.shape[0]
    # Hall of famers in this bin
    num_hof = bin[bin['HoF'] == 1].shape[0]
    
    # Calculate the % of hall of famers in this bin and append to the array.
    pct_of_hof.append(num_hof / num_players)

# Plot the percentage of hall of famers against games played
plt.plot(range(0, max_games, 300), pct_of_hof)
plt.xlabel("Games Played")
plt.ylabel("Percentage of Hall of Famers")
```

As you can see, the percent of all of famers sharply increases around the 2,000 games played mark to over 50% of players being hall of famers! There are 182 game per MLB season so this includes the players who were daily players for over a decade.
<br>
<br>
Now we can observe the pitchers to see if they follow a similar trend but for "outs pitched".

```python
pct_of_hof = []
max_games = df_pitching_career['IPouts'].max()
print(max_games)

for i in range(0, max_games, 3000):
    bin = df_pitching_career[(df_pitching_career['IPouts'] > i) 
                                       & (df_pitching_career['IPouts'] <= i + 3000)]
    
    num_players = bin.shape[0]
    num_hof = bin[bin['HoF'] == 1].shape[0]
    
    pct_of_hof.append(num_hof / num_players)

plt.plot(range(0, max_games, 3000), pct_of_hof)
plt.xlabel("Outs Pitched")
plt.ylabel("Percentage of Hall of Famers")
```

We do indeed see in the graph above there is a very similar shape. Over 50% who have thrown over 10,000 outs (3,333 innings) have been inducted into the hall of fame! These last two visualizations give us confidence that there is a positive relationship between number of games played and chance of being chosen to be in the Hall of Fame.
<br>
<br>
Now that we have established that quantity matters for players, is it also true that quality matters? It is common to define the "best" players as the position players that hit the most home runs or the pitchers who allow the least runs. Below I will visualize a relationship between important statistics per game to see if the best performing players, regardless of number of games played, get inducted into the Hall of Fame.
<br>
<br>
Let's begin with observing hits per at bat by position players. This is similar to the common stat "batting average" and defines what percentage of at bats (times going up to the plate to hit) which result in the player getting a hit! To avoid abnormal errors and edge cases I will only include players who have gotten at least 50 hits in the MLB. It is safe to assume that a player with less than 50 hits is not an MLB Hall of Fame caliber position player. 

```python
# First define a new column to be hits per at bat
df_filtered_hits = df_batting_and_fielding_career[df_batting_and_fielding_career['H'] >= 50]
df_filtered_hits['H_per_AB'] = df_filtered_hits['H'] / df_filtered_hits['AB']

pct_of_hof = []
max_avg = df_filtered_hits['H_per_AB'].max()

# Using np.arange to allow us to use float values for the average
for i in np.arange(0.150, max_avg, 0.03):
    bin = df_filtered_hits[(df_filtered_hits['H_per_AB'] > i) 
                                       & (df_filtered_hits['H_per_AB'] <= i + 0.03)]
    
    num_players = bin.shape[0]
    num_hof = bin[bin['HoF'] == 1].shape[0]
    
    # Avoid division by zero if no players in bin
    if num_players == 0:
        pct_of_hof.append(0)
    else:
        pct_of_hof.append(num_hof / num_players)

plt.plot(np.arange(0.150, max_avg, 0.03), pct_of_hof)
plt.xlabel("Hits per At Bat")
plt.ylabel("Percentage of Hall of Famers")
```

Here is analysis for another batting statistic "RBI" or runs batted in. We also average this statistic per at bat to eliminate quantity as a factor.

```python
# First define a new column to be RBI per at bat
df_filtered_rbi = df_batting_and_fielding_career[df_batting_and_fielding_career['RBI'] >= 10]
df_filtered_rbi['RBI_per_AB'] = df_filtered_rbi['RBI'] / df_filtered_rbi['AB']

pct_of_hof = []
max_rbi = df_filtered_rbi['RBI_per_AB'].max()

# Using np.arange to allow us to use float values for the average
for i in np.arange(0, max_rbi, 0.02):
    bin = df_filtered_rbi[(df_filtered_rbi['RBI_per_AB'] > i) 
                                       & (df_filtered_rbi['RBI_per_AB'] <= i + 0.02)]
    
    num_players = bin.shape[0]
    num_hof = bin[bin['HoF'] == 1].shape[0]
    
    # Avoid division by zero if no players in bin
    if num_players == 0:
        pct_of_hof.append(0)
    else:
        pct_of_hof.append(num_hof / num_players)

plt.plot(np.arange(0, max_rbi, 0.02), pct_of_hof)
plt.xlabel("RBIs per At Bat")
plt.ylabel("Percentage of Hall of Famers")
```

This result is very similar to the last one, we have even more confirmation that these stats correlate to hall of fame induction.

### Hypothesis Testing and Machine Learning
In this section we will use Hypothesis Testing to quantify the extent to which certain statistics relate to hall of fame induction. For contuity we will use some of the same statistics from the previous section. After, we will use Machine Learning to help us make a classifier that can predict future hall of famers based off of their statistics!
<br>
<br>
The first step to hypothesis testing is to select the right test. For this application it's correct to use a Two-Sample T test. This takes two differing samples (HoF vs. not HoF) that you would like to see if there is a difference in values (statistics) of that sample. We can quantify how confident we are that the average statistics of hall of famers exceed those by hall of famers. We can also dial in the value of how much the statistics are exceeded by!
<br>
<br>
In the code we will isolate statistics into two samples, those for hall of famers and those for non hall of famers. A Two-Sample T test will then be performed to check if the mean statistic of HoFers exceeds that of non HoFers by a specified value, 0 if we want to check if it exceeds at all (we suspect this from EDA).

```python
# Import a method to perform a Two-Sample T test from statsmodels module
from statsmodels.stats.weightstats import ttest_ind

# extract the column of career hits by hall of famers
hits_hof = df_batting_and_fielding_career[df_batting_and_fielding_career['HoF'] == 1].H
# extract the column of career hits by non hall of famers
hits_not_hof = df_batting_and_fielding_career[df_batting_and_fielding_career['HoF'] == 0].H

# Perform Two sample T test.
# Provide the two samples
# Difference between means is "larger" than "0"
# "usevar=unequal" to say that we can't assume the variances are equal (they probably aren't)
t, p, df = ttest_ind(hits_hof, hits_not_hof, alternative="larger", usevar="unequal", value=0)
print("We can be {}% confident that the average HoF player has more hits than the average non HoF player"
      .format(100 * (1 - p)))

# Same test but check if the HoF mean is > non-HoF by 2000
t, p, df = ttest_ind(hits_hof, hits_not_hof, alternative="larger", usevar="unequal", value=2000)
print("We can be {}% confident that the average HoF player has 2000 more hits than the average non HoF player\n"
      .format(100 * (1 - p)))

# Same process but for ERA for pitchers
era_hof = df_filtered_era[df_filtered_era['HoF'] == 1].ERA
era_not_hof = df_filtered_era[df_filtered_era['HoF'] == 0].ERA

t, p, df = ttest_ind(era_hof, era_not_hof, alternative="smaller", usevar="unequal", value=0)
print("We can be {}% confident that the average HoF player has a lower ERA than the average non HoF player"
      .format(100 * (1 - p)))

t, p, df = ttest_ind(era_hof, era_not_hof, alternative="smaller", usevar="unequal", value=-0.06)
print("We can be {}% confident that the average HoF player a lower ERA than the average non HoF player by 0.1\n"
      .format(100 * (1 - p)))
```

These are some nice results. Let's continue to the classification stage.
<br><br>
We will use sklearn and a k-NN model to classify write a Hall of Fame classifier. The method is to divide players into testing and training sets using an imported method and train upon those. Each row (player) is labeled as Hof or not in the HoF column which was done in the data curation phase. The machine learning algorithm from sklearn will use these labels to train the model and optimize feature weights to maximize classification accuracy.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import sklearn


target = df_batting_and_fielding_career['HoF']
data = df_batting_and_fielding_career.copy().drop(columns=['playerID', 'HoF', 'POS'])
# Impute all "Na" values to be 0. There are no stats so they have gotten a 0.
data = data.fillna(0)

# Split data 80-20 for training and testing
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size = 0.2)

# Fit a k-NN model with k=7 on the training data
pos_player_model = KNeighborsClassifier(n_neighbors = 5).fit(x_train, y_train)

# Return the accuracy % of this model
print("Accuracy = {}%".format(round(sklearn.metrics.accuracy_score(y_test, pos_player_model.predict(x_test)) * 100, 3)))
```

```python
target = df_pitching_career['HoF']
data = df_pitching_career.copy().drop(columns=['playerID', 'HoF'])
# Impute all "Na" values to be 0. There are no stats so they have gotten a 0.
data = data.fillna(0)

# Split data 80-20 for training and testing
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size = 0.2)

# Fit a k-NN model with k=7 on the training data
pitcher_model = KNeighborsClassifier(n_neighbors = 5).fit(x_train, y_train)

# Return the accuracy % of this model
print("Accuracy = {}%".format(round(sklearn.metrics.accuracy_score(y_test, pitcher_model.predict(x_test)) * 100, 3)))
```

These models have performed great! We have achieved 99% accuracy scores for both the pitcher and position player models! This is generally thought to be a good score, but I'd like to take a bit of a closer look.
<br><br>
Let's examine accuracy for just the hall of famers and just the non-Hall of Famers.

```python
# Give only position players who are in the HoF
pos_player_hof = df_batting_and_fielding_career[df_batting_and_fielding_career['HoF'] == 1]
# Our target column (labels) is binary, whether the player is in the HoF or not.
target = pos_player_hof['HoF']
# Our dataset that we will be predicting from is all quantitative rows that we use as features
data = pos_player_hof.copy().drop(columns=['playerID', 'POS', 'HoF'])
# Use the model to predict based on this data and assess an accuracy score.
pos_player_hof_acc = sklearn.metrics.accuracy_score(target, pos_player_model.predict(data)) * 100

# Repeat for HoF pitchers
pitching_hof = df_pitching_career[df_pitching_career['HoF'] == 1]
target = pitching_hof['HoF']
data = pitching_hof.copy().drop(columns=['playerID', 'HoF'])
pitcher_hof_acc = sklearn.metrics.accuracy_score(target, pitcher_model.predict(data)) * 100

# repeat for non-HoF position players
pos_player_not_hof = df_batting_and_fielding_career[df_batting_and_fielding_career['HoF'] == 0]
target = pos_player_not_hof['HoF']
data = pos_player_not_hof.copy().drop(columns=['playerID', 'POS', 'HoF'])
data = data.fillna(0)
pos_player_not_hof_acc = sklearn.metrics.accuracy_score(target, pos_player_model.predict(data)) * 100

pitching_not_hof = df_pitching_career[df_pitching_career['HoF'] == 0]
target = pitching_not_hof['HoF']
data = pitching_not_hof.copy().drop(columns=['playerID', 'HoF'])
data = data.fillna(0)
pitcher_not_hof_acc = sklearn.metrics.accuracy_score(target, pitcher_model.predict(data)) * 100

print("HoF position players accuracy      = {}%".format(round(pos_player_hof_acc, 3)))
print("HoF pitchers accuracy              = {}%".format(round(pitcher_hof_acc, 3)))
print("Non-HoF position players accuracy  = {}%".format(round(pos_player_not_hof_acc, 3)))
print("Non-HoF pitchers accuracy          = {}%".format(round(pitcher_not_hof_acc, 3)))
```

### Results and Discussion
#### Model Performance:
Wow! These numbers are very surprising, our model is incredibly poor for Hall of Fame players but it is exceptional at predicting Non-HoF players. This shows us that the model is conservative in predicting hall of famers, why may this be?
<br><br>
I believe part of the problem is that there are very many exceptional players that have not been entered into the HoF for a few reasons:<br>
##### 1) 
Inductions are incredibly selective. Somewhere between 2-5 players are inducted each year. There are about 1,000 active players in the MLB at any given time! That means that something has to give, there will be great players that won't be inducted that will throw off our model.<br><br>
##### 2) 
Players must wait 5 years after retirement to be eligble for induction. This means that there are future hall of famers with hall of fame quality stats that are labeled as non-hall of famers in our training set. This mislabeling will throw off our model.<br><br>
##### 3)
Real life hall of fame induction is a human process that takes intangible factors into account when selecting new inductees. Notable MLB writers vote for who they think deserves to be inducted each year. They will take into account (consciously or not) intangibles like leadership, attitude, effort, and communication that is not (and cannot) be quantified in our dataset. The lack of this knowledge will throw off the model and decrease accuracy.<br><br>
For these reasons the model is very conservative in its Hall of Fame award. Repairing these issues are challenging. The only bullet that could be potentially fixed is #2. We could remove all data from players that have not been retired for at least 5 years. The problem with this approach is that we will be erasing decades of data from thousands of players and relying on historic data from over a decade ago. The game of baseball is ever-changing and we would be losing that if we removed data. It's tough to say which approach would yield a better result and is something that could be debated.<br><br>
In conclusion this project was very enlightening for me as a Data Scientist. I mainly gained an insight into the challenges of real-world data. Academic projects almost always are designed such that they work out nicely and for that reason there is an expectation of a great result. When I realized that the model is very conservative and has flaws that is something that was difficult to deal.<br><br>
Overall though, this endeavour was very enlightening and gave me hands on experience with taking a project from start to finish, which is incredibly rewarding.
